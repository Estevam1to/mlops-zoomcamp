Leia os dados de janeiro

Quantas colunas existem?

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

january_df = df[(df['tpep_pickup_datetime'].dt.month == 1) & (df['tpep_dropoff_datetime'].dt.month == 1)]
```

```python
january_df.columns.size
```

Agora vamos calcular a variável de duração. Ela deve conter a duração de uma viagem em minutos.

Qual é o desvio padrão da duração das viagens em janeiro?

```python
# Calculating the duration of the trips in minutes
january_df = january_df.copy()
january_df.loc[:, 'duration'] = (january_df['tpep_dropoff_datetime'] - january_df['tpep_pickup_datetime']).dt.total_seconds()/60
```

```python
january_df
```

```python
std_dev =  january_df['duration'].std()
std_dev 
```

Em seguida, precisamos verificar a distribuição da variável de duração. Existem alguns outliers. Vamos removê-los e manter apenas os registros onde a duração foi entre 1 e 60 minutos (inclusive).

Qual fração dos registros restou após você ter removido os outliers?

```python
other_size = january_df['duration'].size
```

```python
# drop values duration > 60 and duration < 1
january_df = january_df[(january_df['duration'] <= 60) & (january_df['duration'] >= 1)]
january_df
```

```python
new_size = january_df['duration'].size
```

```python
ans = new_size / other_size
ans
```

```python
from sklearn.feature_extraction import DictVectorizer

data_dicts = january_df[['PULocationID', 'DOLocationID']].astype(str).to_dict('records')

vt = DictVectorizer(sparse=True)
vt.fit(data_dicts)

feat = vt.transform(data_dicts)
```

Agora vamos usar a matriz de recursos do passo anterior para treinar um modelo.
- Treine um modelo de regressão linear simples com parâmetros padrão, onde a duração é a variável de resposta
- Calcule o RMSE do modelo nos dados de treinamento


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_taxi = january_df.drop('duration', axis=1)
y_taxi = january_df['duration']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_taxi, y_taxi, test_size=0.2, random_state=0)
```

```python
X_train = vt.transform(X_train[['PULocationID', 'DOLocationID']].astype(str).to_dict('records'))
X_test = vt.transform(X_test[['PULocationID', 'DOLocationID']].astype(str).to_dict('records'))
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
predict = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predict))
rmse
```
